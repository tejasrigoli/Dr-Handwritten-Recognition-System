```python
import numpy as np
import cv2
import matplotlib
import os
import glob
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
#get the path of the images folder
p = r'C:\Users\goli\tejasri\Handwritten-Recognition-System---Major-project>\majorProjectResources\dataset\croppedImages'
path = glob.glob(p+"\*.jpg")
print(path)
```

```python
#read the images
cv_img = []
for img in path:
    n = cv2.imread(img)
    cv_img.append(n)
    print(n.shape)
```

```python
#resize the images to 512x512 pixels
resized_images = []
for img in cv_img:
    resized_img = cv2.resize(img, (512, 512))
    resized_images.append(resized_img)
    #plt.show(resized_img)
    print(resized_img.shape)
```

```python
# to download the resized images in a folder
img = cv2.imread('1.jpg', 1)
path =r'C:\Users\goli\tejasri\Handwritten-Recognition-System---Major-project>\majorProjectResources\dataset\resizedImages'
i=1
for img in resized_images:
    p = str(i)+'.jpg'
    cv2.imwrite(os.path.join(path , p),img)
    i+=1
cv2.waitKey(0)
print("Succesfully saved!!")
```
